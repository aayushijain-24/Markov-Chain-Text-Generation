# Task 3 - Text generation using markov chains
basic statistical model, predicts next word based on previous word(s). no deep learning here, just probability

```python
import random
import re
from collections import defaultdict
```

## sample text
using a small paragraph for now, can replace with any bigger text file later

```python
text = """
deep learning is a subset of machine learning that uses neural networks
with many layers. these models are inspired by the human brain and are
used for tasks like image recognition, natural language processing and
speech recognition. machine learning itself is a subset of artificial
intelligence focused on building systems that learn from data. as more
data becomes available, deep learning models tend to perform better.
"""
```

cleaning the text a bit, lowercase and removing extra symbols

```python
def clean_text(text):
    text = text.lower()
    text = re.sub(r'[^a-z\s]', '', text)
    return text.split()

words = clean_text(text)
print(words[:15])
print("total words:", len(words))
```

## building the markov chain
for each word, store what words come after it (using word pairs / bigrams approach)

```python
def build_markov_chain(words, n=1):
    chain = defaultdict(list)
    for i in range(len(words) - n):
        key = tuple(words[i:i+n])
        next_word = words[i+n]
        chain[key].append(next_word)
    return chain

chain = build_markov_chain(words, n=1)

# just checking how it looks for one key
print(list(chain.items())[:5])
```

## generating text
picking a random starting word, then keep predicting next word based on chain

```python
def generate_text(chain, n=1, length=30):
    start = random.choice(list(chain.keys()))
    result = list(start)

    for _ in range(length - n):
        key = tuple(result[-n:])
        if key not in chain:
            break
        next_word = random.choice(chain[key])
        result.append(next_word)

    return ' '.join(result)

print(generate_text(chain, n=1, length=30))
```

## trying with bigrams (n=2)
using 2 previous words instead of 1, output should make a bit more sense

```python
chain2 = build_markov_chain(words, n=2)
print(generate_text(chain2, n=2, length=30))
```

n=1 gives more random/funny output, n=2 sounds slightly more coherent since it has more context. bigger the input text, better the output will be
